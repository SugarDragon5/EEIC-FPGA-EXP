```python
import matplotlib.pyplot as plt
```

```python
xs=[233,232,184,168,206,179]
ys=[58,134,187,126,69,147]
```

```python
def distance(x1,y1,x2,y2):
    return ((x2-x1)**2+(y2-y1)**2)**0.5
```

```python
d1=distance(xs[0],ys[0],xs[1],ys[1])+distance(xs[1],ys[1],xs[2],ys[2])+distance(xs[3],ys[3],xs[4],ys[4])+distance(xs[4],ys[4],xs[5],ys[5])
d2=distance(xs[0],ys[0],xs[4],ys[4])+distance(xs[4],ys[4],xs[2],ys[2])+distance(xs[3],ys[3],xs[1],ys[1])+distance(xs[1],ys[1],xs[5],ys[5])
print(f"{d1} -> {d2}")
```

```python
plt.clf()
plt.scatter(xs,ys)
plt.plot([xs[0],xs[1],xs[2]],[ys[0],ys[1],ys[2]],'r')
plt.plot([xs[3],xs[4],xs[5]],[ys[3],ys[4],ys[5]],'r')
plt.plot([xs[0],xs[4],xs[2]],[ys[0],ys[4],ys[2]],'b')
plt.plot([xs[3],xs[1],xs[5]],[ys[3],ys[1],ys[5]],'b')
plt.show()
```
